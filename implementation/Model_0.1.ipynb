# Simulation Design

## 0. General Principle

* attributes of the actor contains two kinds of change 
   
    1) "random change" modelling the uncertainty of human being, the magnitude of this kind of change should be small
    
    2) "triggered change" modelling the feedback influence of system upon actor through "Events", like election, ballot etc

* $ A \supseteq A^{E}  \supseteq A^{D} \supseteq A^{I} $ 


1) $A^{E}$: according to motivation, number of candidates should be a parameter of system (k), actors with the highest kth rank of motivation in the population can be nominated as candidates(assign elligibility, a must way to delegator). Only delegators can have input on politial policy

2) $A^{D}$: the flow of delegation is the result of election depending on both socialism(i.g. liquid democracy, representative) and the proximity of actor belief(both current or history)

3) $A^{I}$: Not Implemented

## 1. Actor

### 1.1 Motivation

* Motivation for actors in the socity initialised by power law distribution $ P(X;a)=ax^{a-1}$, $a>0$ and $0<=x<1$ initially
http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.power.html

* **random change of motivation** implemented by adding or substructing a small number coming from a uniform distribution, say [-0.05,0.05], there is no negative motivation or motivation larger than 1, so the motivation stops motivating once touched maximum or minimum (Not Implemented)

* **triggered change of motivation** decided by 1) election  2) ballot (Not Implemented)

* if one actor is elected(delegated), its motivation go positively(say a random number between $[-0.1,0.1]$), if bad things happen, then go negatively 

* if intiative or preferred policy of one actor is elevated in the ballot, its motivation go positively(absolute value of that small number multiple by a big event factor, say 2), otherwise randomly

### 1.2 Belief

* belief of actors in the society on $J_{i,k}$ is initialised by normally distribution $N(0,1)$, belief of policie in for one topic(each actor) is normalized to unit length (called **topic vector**).

* **random changes of belief** happends by tilting the topic vector by a very small degree randomly, chosen from uniform distribution, say $[-\pi/100, \pi/100]$

* if two dimension topic vector the **random change of belief** can be realized by multipling a rotation matrix, if len(topic vector)>2, the current strategy is to tilting only the first two elements

### 1.3 groupId

* indicates which group this actor belongs to, if it is not representative, the groupId for all actors should be all 0

### 1.4 isDel

* int, indicates if one actor is delegator or not (a delegator can have input on the policy), the value of isDel is the weight of its delegation

### 1.5 isEll

* int(0 or 1), indicate whether one candidate is candidate for election 

### 1.6 identification

* integer, position of actor in society.actors list

### 1.7 Motivation, Delegation, Candidate, Ballot History

* actor tracks the history of its own attributes and stored in a list

* even if the non-delegator actors do not take part in ballot, but they do vote for policies "in heart" and record the voting in ballot_history list, this list can be used to calculate BallotHistoryNetwork, a matrix which enables the ballot history of delegators to influence the choice of non-delegator actors in election

## 2. Network

### 2.1 Delegation Network

* network is stored as adjacnet matrix

* a directed network whose edge pointed from the one who has voted to the one who is delegated, for each node(actor) there can/must be one out link

* for representative system, there is no link between **Constituencies**

* the formation of DelegationNetwork depends on BeliefNetwork or BallotHistoryNetwork

* node has in link node should be at least a candidate, not everyone can be candidated in arbitary political system

* to be a candidate or not depend on the motivation of one actor

### 2.2 BeliefNetwork 

* network is stored as adjacnet matrix, whose off diagonal element $a_{ij}$ $i \neq j$ represents the belief proximity between actor i and actor j, whereas the diagonal element is utilized as the storage of elligibility information

* The calculation of BeliefNetwork(BN) depends on the belief attribute of actors 

* BeliefNetwork represents the belief proximity at the current moment

### 2.3 BallotHistoryNetwork

* the same mechanism as BeliefNetwork, but the proximity is calculated by different way using history of actor ballot history list

* BallotHistoryNetwork represents the belief proximity in history, which is recorded at ballot

* allowing BallotHistoryNetwork realizes the system feedback loop  
    belief proximity $\mapsto$ election $\mapsto$ ballot result $\mapsto$ election

## 3 Society Behavior

### 3.1 Election

* voting of actors happens only within group, not across group

* different socialism has different vote method, this should be given as parameter

* procedure

1) election starts by deleting the existing elligibility and delegability

    self._clear_candidates()
    self._clear_delegators()

2) assign candidate by ranking the motivation (motivation is the only criteria influence the assignment of candidate now)

    self._assign_candidates()

3) hold voting to candidates

    votes = self.vote_func(self, m)

4) count voting, assign delegators

    delegators =  self._assign_delegator(votes)

### 3.2 Ballot

* policy decided by the votings from delegators weighted by the number of delegation

* vote for every topic, in each of which the policy possessing the highest number of votes wins

* the vote of delegator has a result on ballot result, non-delegator actor also "vote in heart"

* a run-off ballot could happen when a tie appears (Not Implemented)

### 3.3 legislationPeriod

* delegation network holds unchanged within one legislation period

```python
class Simulator():
    def __init__(self,sim_params):
        """
        :param sim_params: [group_dict,topic_dict,param_dict, socialism_dict]
        """
        self.society = Society(*sim_params)
        #hold first election in new society
        self.society.election()
        self.visualizer = SocietyVisualizer(self.society)
        
    def simul(self,nsteps):
        """Wrapper of society.simul
        """
        start_time = time()
        self.society.simul(nsteps=nsteps)
        end_time = time()
        print "running time: %s seconds" %(end_time-start_time,)
        self.report_basic_information()
        
    
    def report_basic_information(self):
        print "--------------- System Status --------------------"
        print "nsteps: %s " %(self.society.nSteps,)
        print "number of Actors: %s" %(len(self.society.actors),)
        print "Ballot History: %s" %(self.society.system_ballot_history)
        print "delegatiors: %s" %(self.society.delegators)
        print "--------------- Actor Profile --------------------"
        print "actor identification: %s" %self.society.actors[0].idf
        print "actor motivation: %s" %self.society.actors[0].motivation
        print "actor belief: %s" %self.society.actors[0].belief
        print "actor groupId: %s" %self.society.actors[0].groupId
        print "actor delegation status: %s" %self.society.actors[0].isDel
        print "actor candidate status: %s" %self.society.actors[0].isEll
        
        print "actor motivation history: " 
        print self.society.actors[0].motivation_history
        print 
        
        print "actor belief history: " 
        print self.society.actors[0].belief_history
        print 
        
        print "actor ballot history: " 
        print self.society.actors[0].ballot_history
        print
        
        
        
        
    def report_simul_result():
        pass
```
